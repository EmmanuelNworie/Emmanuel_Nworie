First, we set up the notebook environment to use numpy and to display plots inline:

```python
from numpy import *
from matplotlib.pyplot import *
```

We now load our saved data files using the Numpy "loadtxt" routine (hit [shift]-[return] to execute the code block):

```python
x = loadtxt('x.txt')
T = loadtxt('T.txt')
```

We'll first create a simple plot, with the 'x' array defining the horizontal axis points, and the first column of the 'T' array defining the vertical axis points.  Array slicing is nearly identical to Matlab, except that indexing is 0-based, and uses square brackets.  We can add easily add axis labels and a title, by ensuring that the text is enclosed in quotation marks (single or double).  In addition, we can display rendered LaTeX markup by enclosing the text in $-signs:

```python
plot(x,T[:,0])
xlabel('$x$')
ylabel("$y$")
title('$T_1(x)$')
```

Let's go ahead and plot the remaining graphs in separate figure windows as well.  Note that we can concatenate commands with semicolons (as in Matlab and C++), but that we do not need semicolons to end a line:

```python
plot(x,T[:,1]); xlabel('$x$'); ylabel('$y$'); title('$T_3(x)$')
```

```python
plot(x,T[:,2]); xlabel('$x$'); ylabel('$y$'); title('$T_5(x)$')
```

```python
plot(x,T[:,3]); xlabel('$x$'); ylabel('$y$'); title('$T_7(x)$')
```

```python
plot(x,T[:,4]); xlabel('$x$'); ylabel('$y$'); title('$T_9(x)$')
```

Alternatively, we may plot all four functions as subplots within the same figure window:

```python
subplot(3,2,1); plot(x,T[:,0]); xlabel('$x$'); ylabel('$y$'); title('$T_1(x)$')
subplot(3,2,2); plot(x,T[:,1]); xlabel('$x$'); ylabel('$y$'); title('$T_3(x)$')
subplot(3,2,3); plot(x,T[:,2]); xlabel('$x$'); ylabel('$y$'); title('$T_5(x)$')
subplot(3,2,4); plot(x,T[:,3]); xlabel('$x$'); ylabel('$y$'); title('$T_7(x)$')
subplot(3,2,5); plot(x,T[:,4]); xlabel('$x$'); ylabel('$y$'); title('$T_9(x)$')
```

Instead, we can overlay each curve on the same plot by putting each pair of arrays one after another in the plot command, and with any plot we can save the graphic to a file using the "savefig" command:

```python
plot(x,T[:,0],x,T[:,1],x,T[:,2],x,T[:,3],x,T[:,4])
xlabel('$x$')
ylabel('$y$')
title('Chebyshev polynomials')
legend(('$T_1(x)$', '$T_3(x)$', '$T_5(x)$', '$T_7(x)$', '$T_9(x)$'))
savefig('figure1.png')
```

Alternately, since each of the above curves use the same horizontal axis data, and the vertical axis data corresponds to different columns in the matrix T, then the plot command can be shortened considerably (although I am unaware of a shortcut for the legend):

```python
plot(x,T)
xlabel('$x$')
ylabel('$y$')
title('Chebyshev polynomials')
legend(('$T_1(x)$', '$T_3(x)$', '$T_5(x)$', '$T_7(x)$', '$T_9(x)$'))
```

Or we can create the same overlaid plot, with control over line styles and colors, through issuing multiple consecutive plot commands in the same input box; and when saving we can use any of a number of allowable file types (*note: Python is much better at printing directly to PDF than Matlab*):

```python
plot(x, T[:,0], 'b-',  label='$T_1(x)$')
plot(x, T[:,1], 'r--', label='$T_3(x)$')
plot(x, T[:,2], 'm:',  label='$T_5(x)$')
plot(x, T[:,3], 'g-.', label='$T_7(x)$')
plot(x, T[:,4], 'c-', label='$T_9(x)$')
xlabel('$x$')
ylabel('$y$')
title('Chebyshev polynomials')
legend()
savefig('figure2.pdf')
```
